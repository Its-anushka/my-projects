```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Data Preparation and Cleaning**

```python
df = pd.read_csv("heart_disease_health_indicators_BRFSS2015 2 (1) (1).csv")
df
```

**Checking for null values**


```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Sex'] = df['Sex'].map({0: 'Female', 1: 'Male'})
df.head()
```

```python
df['HeartDiseaseorAttack']= df['HeartDiseaseorAttack'].map({1 : "With Heart disease", 0 : "Without Heart disease"})
```

```python
df.head()
```

```python
df['HighBP']= df['HighBP'].map({1 : "High Blood Pressure", 0 : "Normal"})
df['HighChol']= df['HighChol'].map({1 : "High Cholesterol", 0 : "Normal"})
df.head()
```

```python

df['Age'] = df['Age'].map({1: '18-24', 2: '25-30', 3: '30-35', 4: '35-40', 5: '40-45', 6: '45-50', 7:'50-55', 8: '55-60', 9: '60-65', 10: '65-70', 11: '70-75',12: '75-80', 13: '80+' })
df.head()
```

```python
#Categorizing BMI values into categories
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25.0 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

df['Category'] = df['BMI'].map(categorize_bmi)
df['Category']
```

```python
# prompt: Analyze the prevalence of heart disease across different demographic groups (e.g., age, gender, education) using bar charts or pivot tables.

# Analyze the prevalence of heart disease by age group using a bar chart
df1.groupby('Age')['HeartDiseaseorAttack'].value_counts().unstack().plot(kind='bar')
plt.xlabel('Age Group')
plt.ylabel('Number of Cases')
plt.title('Prevalence of Heart Disease by Age Group')
plt.show()


# Analyze the prevalence of heart disease by education level using a bar chart
df1.groupby('Education')['HeartDiseaseorAttack'].value_counts().unstack().plot(kind='bar')
plt.xlabel('Education Level')
plt.ylabel('Number of Cases')
plt.title('Prevalence of Heart Disease by Education Level')
plt.show()


```
